## Line plots

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1, 2, 3, 4])
y = x*2
# first plot with X and Y data
plt.plot(x, y)
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.legend(["blue", "green"], loc="lower right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.array([0,1,2,3])
y1=np.array([3,8,1,10])
x2=np.array([0,1,2,3])
y2=np.array([6,2,7,11])
plt.plot(x1, y1, x2, y2)
plt.legend(["blue", "green"], loc="lower right")
plt.show()
```

## bar plots

```python
import pandas as pd
import matplotlib.pyplot as plt
#read the dataset
tips_data = pd.read_csv("/home/rguktongle/Documents/DS-ML/DS-ML_UDEMY/tips.csv")
#Bar chart againist 
plt.bar(tips_data['total_bill'],tips_data['tip'],color='purple')

plt.title("Bar chart")
#Setting the x and y labels
plt.xlabel("total_bill")
plt.ylabel('tip')

#Adding the legends
plt.show()
```

```python
plt.bar(tips_data['size'],tips_data['tip'])
plt.title("Bar chart")
#Setting the x and y labels
plt.xlabel("size")
plt.ylabel('tip')
plt.legend(['tip'])
#Adding the legends
plt.show()
```

## histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a basic histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

# Display the plot
plt.show()
```
